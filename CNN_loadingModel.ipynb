```python
from tensorflow import keras
model = keras.models.load_model('trained-model/model-recent.h5')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('images/validation/surprise/330.jpg')
plt.imshow(img)
img = cv2.resize(img, (48,48), interpolation = cv2.INTER_LANCZOS4)
img = np.expand_dims(img, 0)
model.predict(img)
```
